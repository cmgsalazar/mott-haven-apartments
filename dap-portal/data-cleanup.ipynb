# DAP Portal

The [Displacement Alert Project (DAP) Portal](https://portal.displacementalert.org/district-dashboard/zipcode/10454) collates information about New York City neighborhoods' housing landscape. For this project, we are targetting only a specific community in the South Bronx that we are identifying under zip code 10454 â€” further filtering this down to only waterfront apartments.

```python
# import libraries
import pandas as pd
```

```python
# read csv

df = pd.read_csv("residential_properties__zipcode10454__12_02_2021-12_09_2024.csv")
df.info()
```

```python
# from https://propertyinformationportal.nyc.gov/

blocks_list = [
    "202319", "202317", "202316", "202309", "202308", "202295", "202296", "202278", "202277", "202261", "202260", "202562", "202546"
]
```

```python
filtered_df = pd.DataFrame()

for block in blocks_list:
    new_df = df[df["BBL"]
    .astype(str)
    .str
    .startswith(block)
    ]

    filtered_df = pd.concat([filtered_df, new_df], ignore_index=True)
```
